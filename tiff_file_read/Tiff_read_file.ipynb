```python
import rasterio
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt
import xarray as xr
```

```python
# Open the file
with rasterio.open('corine.tif') as src:
    print(f'Width: {src.width}, Height: {src.height}')
    print(f'CRS: {src.crs}')
    print(f'Bounds: {src.bounds}')
    print(f'Count: {src.count}')
    print(f'Indexes: {src.indexes}')
    band1 = src.read(1)
    unique_codes = np.unique(band1)
    print(unique_codes)
    src.tags()
    print(src.profile)
    # Go through each band and attempt to print its description
    for i in range(1, src.count + 1):
        description = src.descriptions[i - 1]  # Band descriptions (0-indexed)
        if description:  # Check if the description is not None or empty
            print(f'Band {i} description:', description)
        else:
            print(f'Band {i} has no description in metadata.')
    # Plot the raster data
    plt.imshow(band1, cmap='viridis')



    
    # image = src.read()
    # print(image)
    # show(src)
    
```

```python
# Reading shape file with geopanadas
df = gpd.read_file('2000m.shp.zip')
print (df.head())

df.plot(column='gridcode', legend=True, figsize=(12, 10))
plt.show()

print(f'CRS: {df.crs}')
print(f'Bounds: {df.bounds}')
print(f'Count: {df.count}')
```

```python
polygon = df.loc[2, 'geometry']
gpd.GeoSeries([polygon]).plot()   # GeoSeries to handle geometries, and then plot
plt.show()  # Display the plot
```

```python
# Load the GeoJSON file
gdf = gpd.read_file("BOTTOM_MOUNTED_EUROPE_NUTS0" "_NORWAY_NUTS3.geojson")

# Display the first few rows of the GeoDataFrame
print(gdf.head())

# Check the coordinate reference system of the GeoDataFrame
print(gdf.crs)

# Plot the GeoDataFrame to visualize the data
gdf.plot()

polygon = gdf.loc[0, 'geometry']
gpd.GeoSeries([polygon]).plot()   # GeoSeries to handle geometries, and then plot
plt.show()  # Display the plot
```

```python
# Reading shape file with geopanadas
df = gpd.read_file('zip://WDPA_WDOECM_Jan2024_Public_EU_shp_0.zip')
print (df.head())
```

```python
norway_protected_areas = df[df['ISO3'] == 'NOR']
# Plot the protected areas in Norway
fig, ax = plt.subplots(figsize=(10, 10))
norway_protected_areas.plot(ax=ax, color='green')

# Set title and axis labels (optional)
ax.set_title('Protected Areas in Norway')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# Remove the axes for a cleaner look
ax.set_xticks([])
ax.set_yticks([])

print(norway_protected_areas['DESIG_ENG'])

```
