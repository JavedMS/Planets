```python
import atlite
```

```python
import logging
```

```python
logging.basicConfig(level=logging.INFO)
```

```python
cutout = atlite.Cutout(
    path="europe_2010_V2.nc",
    module="era5",
    x=slice(-13.6913, -1.7712),
    y=slice(49.9096, 60.8479),
    time= slice("2010-01", "2010-02")
)
```

```python
cutout.available_features.to_frame()
```

```python
cutout.prepare(compression={"zlib": True, "complevel": 9})
```
